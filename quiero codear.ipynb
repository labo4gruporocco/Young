```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
```

```python
fig, ax = plt.subplots(figsize=(6, 6))

# Face (Yellow circle)
face = patches.Circle((0.5, 0.5), 0.4, color='#FFD700', ec='#FF8C00', lw=4)
ax.add_patch(face)

# Eyes (White ellipses)
eye_left = patches.Ellipse((0.35, 0.65), 0.12, 0.18, color='white', ec='black', lw=2)
eye_right = patches.Ellipse((0.65, 0.65), 0.12, 0.18, color='white', ec='black', lw=2)
ax.add_patch(eye_left)
ax.add_patch(eye_right)

# Pupils (Cross-eyed looking down)
pupil_left = patches.Circle((0.38, 0.6), 0.04, color='black')
pupil_right = patches.Circle((0.62, 0.6), 0.04, color='black')
ax.add_patch(pupil_left)
ax.add_patch(pupil_right)

# Giant Teardrop
tear = patches.Ellipse((0.38, 0.45), 0.06, 0.15, angle=15, color='#00BFFF')
ax.add_patch(tear)

# Eyebrows 
eyebrow_left = patches.Polygon([[0.25, 0.8], [0.45, 0.75]], closed=False, color='black', lw=3)
eyebrow_right = patches.Polygon([[0.75, 0.8], [0.55, 0.75]], closed=False, color='black', lw=3)
ax.add_patch(eyebrow_left)
ax.add_patch(eyebrow_right)

# Wavy Sad Mouth
x_mouth = np.linspace(0.35, 0.65, 100)
y_mouth = 0.3 + 0.05 * np.sin(x_mouth * 100) - 0.5 * (x_mouth - 0.5)**2  # Wavy + downward curve
ax.plot(x_mouth, y_mouth, color='black', lw=4)

ax.set_xlim(0, 1)
ax.set_ylim(0, 1)
ax.set_aspect('equal')
ax.axis('off')

```
